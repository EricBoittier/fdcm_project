```python
import jinja2
from jinja2 import StrictUndefined

"""
Templates
"""
C_TEMPLATE_STR = open("TEMPLATE_fbuta_c0.sh").read()
C_TEMPLATE = jinja2.Template(C_TEMPLATE_STR, undefined=StrictUndefined)

C_T_TEMPLATE_STR = open("TEMPLATE_t-fbuta_c0.sh").read()
C_T_TEMPLATE = jinja2.Template(C_T_TEMPLATE_STR, undefined=StrictUndefined)

F_TEMPLATE_STR = open("TEMPLATE_fbuta_f0.sh").read()
F_TEMPLATE = jinja2.Template(F_TEMPLATE_STR, undefined=StrictUndefined)

F_T_TEMPLATE_STR = open("TEMPLATE_t-fbuta_f0.sh").read()
F_T_TEMPLATE = jinja2.Template(F_T_TEMPLATE_STR, undefined=StrictUndefined)
```

```python
"""
F Butadiene (cis)
"""

dih = "0_1_2_3"
frames = "/home/unibas/boittier/pydcm-1.2/models/test3/frames.txt"
initial_fit = "/home/unibas/boittier/FDCM/mdcms/f_butadiene/21_charges_refined.xyz"
initial_fit_cube = "/home/unibas/boittier/RDKit_G2/B.pdb/OPT_B3LYP_6-31+G/B"
cubes_dir = "/home/unibas/boittier/RDKit_G2/B.pdb/SCAN_1_2_3_4_S_36_10.0"
scan_name = "_SCAN/B."
n_charges = 21

n_steps_array = [0, 1, 10] #, 100, 1000]

for n_steps in n_steps_array:

    output_dir = f"fbuta-f_{n_steps}"
    _f = F_TEMPLATE.render(dih=dih, n_steps=n_steps, scan_name=scan_name,
                          cubes_dir=cubes_dir, output_dir=output_dir,
                          frames=frames, initial_fit=initial_fit, 
                          initial_fit_cube=initial_fit_cube, n_charges=n_charges)

    f = open(f"../job_files/{output_dir}.sh", "w")
    f.write(_f)
    f.close()

    output_dir = f"fbuta-c_{n_steps}"
    _c = C_TEMPLATE.render(dih=dih, n_steps=n_steps, scan_name=scan_name,
                          cubes_dir=cubes_dir, output_dir=output_dir,
                          frames=frames, initial_fit=initial_fit, 
                          initial_fit_cube=initial_fit_cube, n_charges=n_charges)

    f = open(f"../job_files/{output_dir}.sh", "w")
    f.write(_c)
    f.close()

```
