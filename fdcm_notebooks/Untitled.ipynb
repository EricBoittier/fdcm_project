```python

# for each image, generate a tuple of len==3, representing the image's average RGB value
avg_rgb_values = [(int(x),int(y),int(z)) for x,y,z in zip(df1["a1"], df1["a2"], df1["d1"])]
```

```python
# using the average RGB values, compute the Z-order of each image
hashed_imgs = list(zip(df["frame"],
                   [pymorton.interleave(*avg_rgb) for avg_rgb in avg_rgb_values]))

# returns a sorted-by-color list of photos found within the directory
sorted_list = sorted(hashed_imgs, key=lambda img_tuple: img_tuple[1])

xs = []
ys = []
zs = []
frames = []
for frame, hash_ in sorted_list:
    a, b, c = avg_rgb_values[frame]
    frames.append(frame)
    xs.append(a)
    ys.append(b)
    zs.append(c)

norm = matplotlib.colors.Normalize(vmin=1, vmax=72)
cmap = cm.rainbow
m = cm.ScalarMappable(norm=norm, cmap=cmap)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, ys, zs, c="k", linewidth=0.25)
ax.scatter(xs, ys, zs, ".-", c=m.to_rgba(range(1,73)))
for i, (x,y,z,frame) in enumerate(zip(xs, ys, zs, range(1,73))):
    if i%2 == 0:
        ax.text(x, y, z, f"{frame - i//2}", size=3)
cbar = fig.colorbar(m, ax=ax, orientation="horizontal", pad=0.23)

```
